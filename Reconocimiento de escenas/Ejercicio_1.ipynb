```python
import cv2 as c
import matplotlib.pyplot as plt
```

```python
image = c.imread('sunset.jpg')

imgrgb = c.cvtColor(image, c.COLOR_BGR2RGB)
```

```python
plt.imshow(imgrgb)
```

Escala de grises

```python
img_gris = c.cvtColor(image, c.COLOR_BGR2GRAY)
plt.imshow(img_gris, cmap='gray')
```

```python
bd = c.Canny(img_gris, 100, 150)
plt.imshow(bd, cmap='gray')
plt.title('Deteccin de bordes (Canny)')
plt.axis('off')
plt.show()
```

Mostrando la escala de grises y la de los bordes

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(imgrgb)
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_gris, cmap='gray')
plt.title('Escala de grises')
plt.axis('off')

plt.subplot(1, 3, 3)
bd = c.Canny(img_gris, 100, 200)
plt.imshow(bd, cmap='gray')
plt.title('Deteccin de bordes (Canny)')
plt.axis('off')

plt.show()
```

```python
#Histograma de intensidades
plt.hist(img_gris.ravel(), bins=256, range=[0, 256])
color = ('b')
plt.title('Histograma')
plt.xlabel('Intensidad de pxeles')
plt.ylabel('Frecuencia')
plt.show()
```
